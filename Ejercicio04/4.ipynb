```haskell
procesarLecturas :: (Double -> Double) -> [Double] -> [Double]
procesarLecturas f lecturas = map f lecturas

procesarLecturas (\c -> c * 9/5 + 32) [0, 100, 37]

procesarLecturas (+ 2.5) [10.0, 20.0, 30.0]
```

```haskell
crearFiltro :: (a -> Bool) -> ([a] -> [a])
crearFiltro predicado = filter predicado

soloPares = crearFiltro even
soloPositivos = crearFiltro (> 0)
soloLargos = crearFiltro (\s -> length s > 4)

print (soloPares [1..10])               
print (soloPositivos [-3,-1,0,2,5])     
print (soloLargos ["ana","pedro","juan","alejandro"]) 
```

```haskell
enRango :: Int -> Int -> (Int -> Bool)
enRango min max = (\x -> x >= min && x <= max)

entre5y10  = enRango 5 10
esMenorEdad = enRango 0 17
esAdultoMayor = enRango 65 120

print (entre5y10 7)          
print (entre5y10 11)         
print (esMenorEdad 15)       
print (filter (enRango 5 10) [1..15])  
```

```haskell
potenciador :: Int -> (Int -> Int)
potenciador n = (\x -> x ^ n)

cuadrado = potenciador 2
cubo     = potenciador 3

print (cuadrado 5)             
print (cubo 3)                 
print (map (potenciador 2) [1..5])
print (map (potenciador 3) [1..5])  
```
